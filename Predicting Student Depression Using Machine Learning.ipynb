# **Machine Learning**

# **Import Library Function For ML**

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix, precision_score, recall_score, f1_score
from sklearn.svm import SVC
```

# **Splitting the datasets for ML**

**Train Data**

```python
X_train, X_test, Y_train, Y_test = train_test_split(df.drop('Depression', axis=1), df['Depression'], test_size=0.80, random_state=45)
```

```python
X_train.head()
```

```python
Y_train.head()
```

```python
X_test
```

```python
Y_test
```

# **Apply Logistic Regression Algorithm**

```python
reg = LogisticRegression()
reg.fit(X_train, Y_train)
```

```python
Ypred = reg.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score, confusion_matrix
```

```python
accuracy = accuracy_score(Y_test, Ypred)*100
accuracy
```

```python
precision = precision_score(Y_test, Ypred)*100
precision
```

```python
recall = recall_score(Y_test, Ypred)*100
recall
```

```python
f1 = f1_score(Y_test, Ypred)*100
f1
```

```python
confusion_matrix = metrics.confusion_matrix(Y_test,Ypred)
cm_display = metrics.ConfusionMatrixDisplay(confusion_matrix = confusion_matrix, display_labels = [True, False])
cm_display.plot()
plt.show()
```

# **Applying Support Vector Machine**

```python
model = SVC(kernel='linear', random_state=42)
model.fit(X_train, Y_train)
```

```python
Y_pred = model.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score
```

```python
accuracy = accuracy_score(Y_test, Y_pred)*100
print("Accuracy:", accuracy)
```

```python
precision = precision_score(Y_test, Y_pred)*100
print("Precision:", precision)
```

```python
recall = recall_score(Y_test, Y_pred)*100
print("Recall (Sensitivity):", recall)
```

```python
f1 = f1_score(Y_test, Y_pred)*100
print("F1-Score:", f1)
```

```python
confusion_matrix = metrics.confusion_matrix(Y_test,Y_pred)
cm_display = metrics.ConfusionMatrixDisplay(confusion_matrix = confusion_matrix, display_labels = ['True', 'False'])
cm_display.plot()
plt.show()
```
